```python
import numpy as np
import matplotlib.pyplot as plt

def lorentz_curve(sigma, r, b, tmax=100, dt=0.01):
    # Initial conditions
    x, y, z = 0.0, 1.0, 1.05
    # Time array
    t = np.arange(0, tmax, dt)
    # Arrays to store the x, y, z values
    xs = np.empty(len(t))
    ys = np.empty(len(t))
    zs = np.empty(len(t))

    for i in range(len(t)):
        xs[i] = x
        ys[i] = y
        zs[i] = z
        # Derivatives
        dx = sigma * (y - x)
        dy = x * (r - z) - y
        dz = x * y - b * z
        # Update the variables
        x += dx * dt
        y += dy * dt
        z += dz * dt

    return xs, ys, zs

# Parameters
sigma = 10.0
r = 28.0
b = 8.0 / 3.0

# Generate Lorentz curve data
x, y, z = lorentz_curve(sigma, r, b)

# Plotting
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot(x, y, z, lw=0.5)

# Labels
ax.set_xlabel("X Axis")
ax.set_ylabel("Y Axis")
ax.set_zlabel("Z Axis")

# Title
ax.set_title("Lorentz Attractor")

# Show plot
plt.show()
```
